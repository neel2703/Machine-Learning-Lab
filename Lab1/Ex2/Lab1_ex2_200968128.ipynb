```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
```

# 1. Use the boston.csv dataset and determine the best 5 features to predict ‘MEDV’.

```python
df = pd.read_csv('Boston.csv')
df.head()
```

```python
corrmedv = df.corr()
top_corr = corrmedv.index
plt.figure(figsize=(10,10))
sns.heatmap(df[top_corr].corr(), annot=True, cmap="RdYlGn")
```

```python
#we can see that 'zn', 'rm', 'chas', 'dis' and 'black' are the features that have the highest correlation with 'medv' and are the best 5 features to predict 'medv'.
x = df[['zn', 'rm', 'black']]
y = df[['medv']]
```

# 2. Using sklearn.linear_model, find the multiple regression model for the boston.csv dataset using the best 3 features.

```python
#the best 3 features are 'rm', 'zn', and 'black'.

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2)
linear_reg = LinearRegression().fit(x_train, y_train)
y_pred = linear_reg.predict(x_test)
```

# 3. Find the accuracy of the model using appropriate metrics using 80, 20 split for training and test.

```python
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

r2s = metrics.r2_score(y_test, y_pred)

print('R2 score = ',r2s)
print('Root Mean Squared Error = ', rmse)
```
